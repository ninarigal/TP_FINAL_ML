```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
```

```python
file_path = r'/Users/ninarigal/UDESA/TERCER AÑO/APRENDIZAJE AUTOMÁTICO Y APRENDIZAJE AUTÓNOMO/TP_FINAL_ML/pf_suvs_i302_1s2024.csv'
data = pd.read_csv(file_path)
```

```python
current_year = 2024
data['Edad'] = current_year - data['Año']
data = data.drop(columns=['Año', 'Color', 'Tipo de combustible', 'Motor', 'Tipo de carrocería', 'Título', 'Tipo de vendedor', 'Con cámara de retroceso', 'Puertas', 'Transmisión'])
data = data.drop(columns='Versión')
```

```python
def pesos_dolares(x):
    for i in range(len(x)):
        if x['Moneda'][i] == '$':
            x.loc[i, 'Precio'] = x.loc[i, 'Precio'] / 1040
            # x['Precio'][i] = x['Precio'][i] / 1040
    return x
```

```python
pesos_dolares(data)
data = data.drop(columns='Moneda')
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_marca_edad(data, marca):
    data_marca = data[data['Marca'] == marca]
    plt.hist(data_marca['Edad'], bins=20)
    plt.title(marca)
    plt.show()
```

```python
for marca in data['Marca'].unique():
    plot_marca_edad(data, marca)
```

```python
def plot_edad_precio_modelo(data, modelo):
    data_modelo = data[data['Modelo'] == modelo]
    plt.scatter(data_modelo['Edad'], data_modelo['Precio'])
    plt.xlabel('Edad')
    plt.ylabel('Precio')
    plt.title(modelo)
    plt.show()
    
```
