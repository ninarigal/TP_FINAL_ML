```python
import re

# Definir la función para limpiar las versiones
def clean_version(version):
    cleaned_version = " ".join(str(version).split()).strip().lower()
    #si dentro de una palabra hay una tilde la reemplazo por la vocal sin tilde
    cleaned_version = re.sub(r'[á]', 'a', cleaned_version)
    cleaned_version = re.sub(r'[é]', 'e', cleaned_version)
    cleaned_version = re.sub(r'[í]', 'i', cleaned_version)
    cleaned_version = re.sub(r'[ó]', 'o', cleaned_version)
    cleaned_version = re.sub(r'[ú]', 'u', cleaned_version)
    return cleaned_version

# Hago una copia de la columna 'Titulo' para ver si la función funciona correctamente
data['Título versión'] = data['Título'].apply(clean_version).str.cat(data['Versión'].apply(clean_version), sep=' ')

# data['Título'] = data['Título'].apply(clean_version)
```

PEUGEOT

```python
keywords = ['active', 'allure', 'feline', 'thp', 'concert', 'crossway', 'gt', 'hybrid']

# Definir la función de mapeo para agrupar versiones similares
def map_version(version):
    for keyword in keywords:
        if re.search(rf'\b{keyword}\b', version):
            return keyword
    return 'other'

# Filtrar solo las filas de la marca 'peugeot'
data_peugeot = data[data['Marca'] == 'peugeot'].copy()

# Aplicar la función de mapeo a la columna 'Versión'
data_peugeot['Título versión'] = data_peugeot['Título versión'].apply(map_version)

# Si quieres actualizar el DataFrame original
data.update(data_peugeot)
```

TOYOTA

```python
data_toyota = data[data['Marca'] == 'toyota'].copy()

# si dice grs, gr-sport, gazoo racing, gr-s, gr
# si dice diamons, diamante, diamond
# si dice prado, vx -> vx

# Definir las palabras clave que deseas buscar y reemplazar
grs = ['gr-sport', 'grs', 'gazoo racing', 'gr-s']

# Función para reemplazar las palabras clave
def replace_grs(version):
    for keyword in grs:
        if keyword in version:
            return 'gr'
    return version

# Aplicar la función a la columna 'Título versión'
data_toyota['Título versión'] = data_toyota['Título versión'].apply(replace_grs)

diamond = ['diamons', 'diamante']

# Función para reemplazar las palabras clave
def replace_diamond(version):
    for keyword in diamond:
        if keyword in version:
            return 'diamond'
    return version

# Aplicar la función a la columna 'Título versión'
data_toyota['Título versión'] = data_toyota['Título versión'].apply(replace_diamond)

vx = ['prado']

# Función para reemplazar las palabras clave
def replace_vx(version):
    for keyword in vx:
        if keyword in version:
            return 'vx'
    return version

# Aplicar la función a la columna 'Título versión'
data_toyota['Título versión'] = data_toyota['Título versión'].apply(replace_vx)

#rav4 y at
# Función para reemplazar las palabras clave
def replace_rav4_at(version):
    if 'at' in version:
        return 'rav4 at'
    return version

def replace_rav4_cvt(version):
    if 'cvt' in version:
        return 'rav4 cvt'
    return version

# Aplicar la función a la columna 'Título versión'
index_rav4 = data_toyota[data_toyota['Modelo'] == 'RAV4'].index
data_toyota.loc[index_rav4, 'Título versión'] = data_toyota.loc[index_rav4, 'Título versión'].apply(replace_rav4_at).apply(replace_rav4_cvt)

keywords = ['d', 'limited', 'hev', 'seg', 'xli', 'xei', 'gr', 'srx', 'sr', 'srv', 'runner', 'diamond', 'vx', 'v8', 'dl', 'tx', 'xle', 'rav4 at', 'rav4 cvt']

# Aplicar la función de mapeo a la columna 'Versión'
data_toyota['Título versión'] = data_toyota['Título versión'].apply(map_version)

# Imprimo los que no se mapearon
data_toyota_other = data_toyota[data_toyota['Título versión'] == 'other']
for i in range(len(data_toyota_other)):
    print(data_toyota_other['Marca'].iloc[i], data_toyota_other['Modelo'].iloc[i], data_toyota_other['Título'].iloc[i], data_toyota_other['Versión'].iloc[i])

# Si quieres actualizar el DataFrame original
data.update(data_toyota)
```

```python
data_citroen = data[data['Marca'] == 'citroën'].copy()

keywords = ['shine', 'thp', 'first', 'feel', 'tendance', 'feel', 'live', 'noir', 'puretech', 'series', 'origins']

#firts a first
#tendence a tendance
#fell a feel

# Función para reemplazar las palabras clave
def replace_first(version):
    if 'firts' in version:
        return 'first'
    return version

# Función para reemplazar las palabras clave
def replace_tendance(version):
    if 'tendence' in version:
        return 'tendance'
    return version

# Función para reemplazar las palabras clave
def replace_feel(version):
    if 'fell' in version:
        return 'feel'
    return version

# Aplicar la función a la columna 'Título versión'
data_citroen['Título versión'] = data_citroen['Título versión'].apply(replace_first).apply(replace_tendance).apply(replace_feel)

# Aplicar la función de mapeo a la columna 'Versión'
data_citroen['Título versión'] = data_citroen['Título versión'].apply(map_version)

# Imprimo los que no se mapearon
data_citroen_other = data_citroen[data_citroen['Título versión'] == 'other']
for i in range(len(data_citroen_other)):
    print(data_citroen_other['Marca'].iloc[i], data_citroen_other['Modelo'].iloc[i], data_citroen_other['Título'].iloc[i], data_citroen_other['Versión'].iloc[i])

# Si quieres actualizar el DataFrame original
data.update(data_citroen)
```
