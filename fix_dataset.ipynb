```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
```

```python
file_path = r'pf_suvs_i302_1s2024.csv'
data = pd.read_csv(file_path)
```

```python
current_year = 2024
data['Edad'] = current_year - data['Año']
data = data.drop(columns=['Año']) # Drop the original year column (we already have the age)
data = data.drop(columns=[ 'Tipo de combustible', 'Tipo de carrocería', 'Con cámara de retroceso', 'Puertas'])
# Drop the columns that are not useful for the model

print(data.columns)
```

```python
# Fix the column 'Transmisión'
data['Transmisión'] = data['Transmisión'].replace('Semiautomática', 'Automática secuencial')
```

Ver que hacer con los 32 que no especifican

```python
print(data['Transmisión'].unique())
print(data['Transmisión'].isnull().sum()) #Hay 32 autos sin transmisión especificada
for i in range(len(data)):
    if pd.isnull(data['Transmisión'][i]):
        print(data['Marca'][i], data['Modelo'][i], data['Título'][i], data['Versión'][i])
```

```python
def pesos_dolares(x):
    for i in range(len(x)):
        if x['Moneda'][i] == '$':
            x.loc[i, 'Precio'] = x.loc[i, 'Precio'] / 1040 # 1 peso = 1040 dolars (9th may 2024)
    return x
```

```python
pesos_dolares(data)
data = data.drop(columns='Moneda') # Drop the column 'Moneda' since we already converted the prices to dollars (is not useful anymore)
```
