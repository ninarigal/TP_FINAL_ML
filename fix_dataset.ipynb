```python
import re

# Definir la función para limpiar las versiones
def clean_version(version):
    cleaned_version = " ".join(str(version).split()).strip().lower()
    #si dentro de una palabra hay una tilde la reemplazo por la vocal sin tilde
    cleaned_version = re.sub(r'[á]', 'a', cleaned_version)
    cleaned_version = re.sub(r'[é]', 'e', cleaned_version)
    cleaned_version = re.sub(r'[í]', 'i', cleaned_version)
    cleaned_version = re.sub(r'[ó]', 'o', cleaned_version)
    cleaned_version = re.sub(r'[ú]', 'u', cleaned_version)
    return cleaned_version

data['Versión'] = data['Versión'].apply(clean_version)
```

```python
keywords = ['active', 'allure', 'feline', 'thp', 'concert', 'crossway', 'gt', 'hybrid']

# Definir la función de mapeo para agrupar versiones similares
def map_version(version):
    for keyword in keywords:
        if re.search(rf'\b{keyword}\b', version):
            return keyword
    return 'other'

# Filtrar solo las filas de la marca 'peugeot'
data_peugeot = data[data['Marca'] == 'peugeot'].copy()

# Aplicar la función de mapeo a la columna 'Versión'
data_peugeot['Versión'] = data_peugeot['Versión'].apply(map_version)

# Si quieres actualizar el DataFrame original
data.update(data_peugeot)
```
